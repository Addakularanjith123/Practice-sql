```python
#if we want to print the age of the movies(means current year(2026)-release year)
df['age']=df['release_year'].apply(lambda x: 2026 - x) #here the release_year is performing operation on the column
                                                      # the lambda function is like a for loop (performing operation on the each row)
```

```python
#In the above we have only one existing column
```

```python
df.head(4) #new column has been added at the end of the dataframe
```

```python
# what we perform operation on 2 existing columns like profit of the movie then to find it require two columns (revenue-budget column)
df['profit']=df.apply(lambda x: x['revenue'] - x['budget'] ,axis=1) #when axis=1 that means change in the column
```
