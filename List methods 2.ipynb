```python
# 3. List Reference Behavior
# When you assign one list to another variable,
# both refer to the same memory location
x = [1, 2, 3]
y = x

# Modifying y will also affect x
y.append(4)

# Output will be [1, 2, 3, 4]
print(x)


# 4. List Methods Demonstration

# Initial list
n = [1, 2, 3, 8, 9, 10]
print(n)

# remove() -> removes element by value (not index)
# If value not found, it throws an error
n.remove(2)
print(n)

# pop() -> removes element by index
# Here removing element at index 4
n.pop(4)
print(n)

# append() -> adds a single element at the end
n.append(4)
print(n)

# extend() -> adds multiple elements at the end
n.extend([5, 6, 7, 11, 12])
print(n)

# clear() -> removes all elements from the list
n.clear()
print(n)


# 5. pop() Method Example

# Original list
my_list = [10, 20, 30, 40, 50]
print(f"Original list: {my_list}")

# pop(index) removes and returns the element
removed_element = my_list.pop(2)

# Updated list after removal
print(f"List after removing element at index 2: {my_list}")

# The removed element
print(f"Removed element: {removed_element}")
```
