converting n.a or not available values into NaN(Proper NA) values using dictionary like specified columns

```python
# the above values n.a and not available values are strings
df = pd.read_csv("/content/drive/MyDrive/All files/stock_data.csv", header=1,na_values={
    'eps':['not available']
})
```

```python
df #the eps column and row RIL now converted to NaN value
```

```python
df = pd.read_csv("/content/drive/MyDrive/All files/stock_data.csv", header=1,na_values={
    'eps':['not available'],
    'revenue':[-1]
})
df #the -1 value converted into NaN
```
