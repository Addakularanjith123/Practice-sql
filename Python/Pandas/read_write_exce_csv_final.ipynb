```python
import numpy as np

def safe_divide(x, y):
    if pd.isna(x) or pd.isna(y):
        return np.nan
    return np.divide(float(x),float(y))
```

```python
df["pe"] = df.apply(lambda x: safe_divide(x["price"], x["eps"]), axis=1)
df
```

```python
#raw metrix are like(columns i.e 'eps','revenue','price'...) to derived metrix (like derived column)
```

```python
#To get price to earning ration then we do price/eps
df['pe']=df['price']/df['eps']
```

**Write two dataframes to two separate sheets in excel**

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```
