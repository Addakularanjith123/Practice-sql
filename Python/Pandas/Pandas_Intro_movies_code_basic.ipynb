We have movies dataset. We want to do some basic analysis such as,

1. What is the min, max and average rating for all movies?
1. What is the min, max and average rating for all bollywood movies?
1. What is the min, max and average rating for all hollywood movies?

We will see how this can be done first without using Pandas and then using Pandas

## Option 1 - Movies analysis without using Pandas

```python
import csv

def calculate_rating_stats(data, industry=None): #Here industry as the parameter and default value is None
                                            # the statistics for the entire data set and individual statistics for bollywood and hollywood industries
    ratings = [] #It will store the valid list of elements only
    for row in data:
        if row[3]!='NULL' and (not industry or row[1]==industry):#In every row the 4th column(rating column) is not equal to NULL and
              ratings.append(float(row[3]))  # conveting into the float to perform the operations
              '''
                  Industry = "Bollywood"	Only Bollywood rows included#In every row the 4th column(rating column) is not equal to NULL and
          row[3] != 'NULL'
            Checks if the rating (4th column) is not 'NULL' This avoids invalid/missing values

            👉 Example:

            '8.5' → valid ✔️
            'NULL' → skipped ❌

      (not industry or row[1] == industry)

          This part controls filtering:

          not industry → True when no industry is provided
          row[1] == industry → True when row matches the given industry

          👉 So overall:

          Scenario	                                           Result
          No industry                                   given	All rows included
          Industry = "Bollywood"	                      Only Bollywood rows included

               '''

    max_rating = max(ratings)
    min_rating = min(ratings)
    avg_rating = sum(ratings) / len(ratings)
    return max_rating, min_rating, avg_rating


with open('movies.csv') as f:
    data = list(csv.reader(f)) #list of rows (each row is a list)
    header = data[0]  # The row 0 considered has header in the file
    data = data[1:]  #actual data is starting from the row 1st

    #Printing the statistics

    max_rating, min_rating, avg_rating = calculate_rating_stats(data) # It will do calculate the statistics for the entire date because there is no another argument pass
    print(f"All records: Min rating = {min_rating}, Max rating = {max_rating}, Avg rating = {avg_rating}")

    max_rating, min_rating, avg_rating = calculate_rating_stats(data, industry="Bollywood") #It will do calulate the statistics for the only Bollywood because the another argument has been passed
    print(f"Bollywood: Min rating = {min_rating}, Max rating = {max_rating}, Avg rating = {avg_rating}")

    max_rating, min_rating, avg_rating = calculate_rating_stats(data, industry="Hollywood")  #It will do calulate the statistics for the only Hollywood because the another argument has been passed
    print(f"Hollywood: Min rating = {min_rating}, Max rating = {max_rating}, Avg rating = {avg_rating}")
```
