**Replacing single value**

```python
df.replace(-99999,1) #only -99999 value is replacing with the 1.
```

```python
df #because there is no inplace=True
```

**Replacing list with nan value**


Numpy is a library and pandas uses under it.

```python
import numpy as np
df.replace(-99999,np.nan) #replacing the -99999 value with NaN we need to use numpy library
```

```python
df
```

Explicitly Mensioning

```python
df.replace(-99999,value=np.nan) #both are same just used the value key word argument
```

**Replacing list with single value**





```python
df.replace(to_replace=[-99999,-88888],value=0) #replacing the list of values with 0 with the key word aruments.
```

```python
# df.replace(to_replace=[-99999,-88888],1) #single to_place argument won't work but single value key word argument will work
```

```python
df
```

```python
df.replace([-99999,-88888],np.nan) #both values are replacing with NaN
```

```python
df
```

```python
df.replace(to_replace=[-99999,-88888],value=np.nan) #list of values are replaced with NaN with key word arguments
```

```python
df
```

**Replacing per column**

```python
#we have to mention the column name as like a dictionary keys and value is replacing value(like nan,1,0)
df_new=df.replace({
    'temperature':-99999,
    'windspeed':[-99999,-88888],
    'event':'no event'
},np.nan)
df_new
```

```python
df
```

**Replacing by using mapping**

```python
df.replace({
    -99999:np.nan #for all the -99999 in every column it will replace with NaN(observe row 3 and 4 in temparature and windspeed columns)
})
```

```python
df
```

```python
df.replace({
    -99999:1, #replacing with 1 for all the column values who has value -99999
    -88888:np.nan, #for all -88888 values it will replace with NaN
    'no event':'Rain' #considering the no event value as Rain
})
```

**Replacing list with another list**

```python
#creating a new data frame key are column names and list is values the length has to be same
df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df
```
