```python
df = pd.read_csv("/content/drive/MyDrive/All files/weather_data.csv",parse_dates=['day'])
df.set_index('day',inplace=True)
```

```python
df #to fill NaN value with another one(like mean or median or anything)
```

```python
# from google.colab import sheets
# sheet = sheets.InteractiveSheet(df=df)
```

```python
df_tm=df.temperature.mean() #to konw the mean of the entire column
df_w=df.windspeed.median() #median of the windspeed column
df_wm=df.windspeed.mode() #No mode because no number repeates
```

```python
print(df_tm) #33.22
print(df_w) #8.1
print(df_wm) #0     6.0
            # 1     7.0
            # 2     8.1
            # 3     9.0
            # 4    12.0
```

<font color="pink">**Fill na using column names and dict**</font>

```python
df
```

```python
df.fillna({
    'temperature':df.temperature.mean(),
    'windspeed':df.windspeed.median()
})
```

```python
new_df=df.fillna({  #key is the column name in dataframe
    'temperature':df.temperature.mean(),#replacing Nan values with mean i.e(33.22)
    'windspeed':df.windspeed.median(),  #replacing Nan values with median i.e(8.1) #better to use mean because if we have outlier the it won't calculat(1,2,3,2,1001 now mean is 201 and median is 2 middle value asc or desc order)
    'event':'No Event'  #replacing Nan values with "No Event"
})
```

```python
new_df #another dataframe is created
```

```python
df
```

<font color="purple">**Use method to determine how to fill na values**</font>

```python
new_df_f = df.fillna(method="ffill") #forward fill in the NaN cell it will replace the before cell value
new_df_f
```

```python
# dir(df)
```

```python
new_df_b=df.fillna(method='bfill')   #backward fill in the NaN cell it will replace the after(next) cell value
new_df_b
```

```python
df
```

<font color="purple">**Use of axis**</font>

```python
new_df_b = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns" #Replacing NaN values with the column values
new_df_b     #in day column 2017-01-5  in event column 'Snow' value is replace the windspeed column value i.e from NaN to 'Snow'
```

```python
new_df_f = df.fillna(method="ffill", axis="columns")
new_df_f #In day column 2017-01-07	the windspeed column has been changed to temparature cell value

#there are no value in the both  values with NaN or not even single cell in wind speed or event column with NaN value with the value in the temparature column to fill
```

```python
df
```

<font color="purple">**limit parameter**</font>

```python
new_df=df.fillna(method='ffill',limit=1) #If we have no.of NaN value followed by only one value then it will replace only next(limit) NaN value with the valid value
new_df            #we can observe day 07 and 08 only 07 is changed but not the 08 with the value because of the limit 1
```

### <font color="blue">dropna</font>

```python
df.dropna() #any row even if the single NaN it will drop the entire row
```

```python
df
```

```python
df.dropna(how='any') #even single NaN drop the entire row
```

```python
df.dropna(how="all") #if all column in a row with NaN value then only it will drop the row
                     # day 09 column has droped because the entire row is NaN values
```

```python
df
```

```python
df.dropna(thresh=2) #If any 2 NaN values in a row then it will drop
                    #day 07 and 08 rows are droped because it has 2 NaN values in a row.
```
