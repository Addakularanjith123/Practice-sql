```
#2.Fast Access
l1=[1,2,3]
l2=[8,9,10]
#If we want to perform addition i.e 1st element of l1 add it to the 1st element of l2
print(l1+l2)
print(zip(l1,l2)) #Zip object
print(tuple(zip(l1,l2))) #It converted to tuple and printing
for x,y in zip(l1,l2):
  print(x,y)
```

```
size=1000000
l1=list(range(size))
l2=list(range(size))
l3=[x+y for x,y in zip(l1,l2)] #list comprehenssion same like the above one
```

```
# measure the performance of list
#Performance checking of the list by using time
import time
start_time=time.time()
size=1000000
l1=list(range(size))
l2=list(range(size))
l3=[x+y for x,y in zip(l1,l2)]
end_time=time.time()
print("python list took to perform:",end_time-start_time)
```

```
#Measure the performance of Numpy
size=1000000
n1=np.arange(size)
n2=np.arange(size)
n1[:5],n2[:5]
```

**It will perform vectorized operation**

```
start_time=time.time()
n1+n2
end_time=time.time()
print("python numpy took to perform:",end_time-start_time)
```
