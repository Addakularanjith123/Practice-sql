## Indexing, Slicing

```python
a = np.array([6,7,8,9])
a[0:2] #all the element from 0(inclusive) to 2(exclusive)
```

```python
a[-4:-2] #from the -4 indexes to -2 index it's like a backword index
```

```python
a[-4:2] #same but forward index (Here we have specified the index)
```

```python
a[-3:] #from -3 index till end because there is no end index
```

```python
#If we want to print last 2 elements
a[2:] #colon and then nothing it means all the remaining elements
```

```python
#If we want to print last 2 elements another way
a[-2:]
```

```python
a[-2:3] #-2 is the starting index i.e 8 and upto index 3(exclusive) i.e(6,7,8,9) therefore it will satisfy only the -2 index element i.e 8
```

```python
a[-2:4] #this condition satisfy the last 2 elements
```

```python
import numpy as np

transactions = np.array([
    [101, 'Mohan', 250.50, '2023-08-01'],
    [102, 'Bob', 150.00, '2023-08-02'],
    [103, 'Fatima', 300.75, '2023-08-01'],
    [104, 'David', 400.20, '2023-08-03'],
    [105, 'Aryan', 330.1, '2023-08-04'],
])

# Build new rows
result = np.array([
    [row[0], row[1][:1], row[1][1:], row[2], row[3]]
    for row in transactions #here we are spliting col 1 twice in the every row
])

print(result) #row[1][:1]   # 'M' row[1][1:]   # 'ohan'
```

```python
for row in transactions: #there is no allign
    print(f"{row[0]} {row[1][:1]} {row[1][1:]} {row[2]} {row[3]}")
```

### Boolean operations

```python
#To find in which month we went for the below the target
monthly_sales = np.array([30, 33, 35, 28, 42]) #consider('Jan','Feb',.....)

result = monthly_sales < 32 #if the condition satisfies then it gives true value else false values(i.e boolean values)
result
```

```python
print(monthly_sales[result]) #To know the result value
```

```python
monthly_sales[result] = -1 #replacing when the condition true with -1 value
monthly_sales
```

In monthly if we want to find out the maximum sales

```python
monthly_sales.max() #just knowing the value not the index(like which month has highest sales)
```

#Argmax

```python
index #in the 4th index the maximum value is present
```
