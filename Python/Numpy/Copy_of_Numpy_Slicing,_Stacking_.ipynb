## Indexing, Slicing

```python
a = np.array([6,7,8,9])
a[0:2] #all the element from 0(inclusive) to 2(exclusive)
```

```python
a[-4:-2] #from the -4 indexes to -2 index it's like a backword index
```

```python
a[-4:2] #same but forward index (Here we have specified the index)
```

```python
a[-3:] #from -3 index till end because there is no end index
```

```python
#If we want to print last 2 elements
a[2:] #colon and then nothing it means all the remaining elements
```

```python
#If we want to print last 2 elements another way
a[-2:]
```

```python
a[-2:3] #-2 is the starting index i.e 8 and upto index 3(exclusive) i.e(6,7,8,9) therefore it will satisfy only the -2 index element i.e 8
```

```python
a[-2:4] #this condition satisfy the last 2 elements
```

#If some information has in one array and remaining information in another array


```python
# Customer ID, Name
c = np.array([
    [101, 'Mira'],
    [102, 'Abdul'],
    [103, 'Andrea']
])
# Customer Id, Purchase Amount, Purchase Date
d = np.array([
    [101, 250.50, '2023-08-01'],
    [102, 150.00, '2023-08-02'],
    [103, 300.75, '2023-08-01']
])
```

#vstack,hstack

##Creating a combine array

```python
# If we need some information in one array and the remaining information from another array
np.hstack((c,d)) #Horizontal stacking just putting 2 boxes together horizontally,we have to specify as tuple.

# Here the 1st col(101,102,103) has been repeated it is not doing join(merge will discuss in pandas)
```
