### Product and geographic breakdown of Apple
Product wise breakdown of the apple sales in different countries in quarter 1 and quarter

Screenshot 2026-04-13 131601.png

We just convert the above image into numpy array

```python
import numpy as np
```

```python
# Sales data for Quarter 1 (Matrix 1)
# Rows represent different products, columns represent different regions

#It is 2D Array (Best way to represent in matrix)

q1 = np.array([
    [200, 220, 250],  # Product A
    [150, 180, 210],  # Product B
    [300, 330, 360]   # Product C
])

q2 = np.array([
    [209, 231, 259],  # Product A
    [155, 192, 222],  # Product B
    [310, 340, 375]   # Product C
])
```

```python
#I want to combine the sales numbers of q1 and q2
q1+q2 #combinely of q1 and q2 units
```

```python
# Total sales
q1_sales = np.sum(q1)   # (200+220+250+150+180+210+300+330+360)
q2_sales = np.sum(q2)

print("Q1 Sales:", q1_sales)
print("Q2 Sales:", q2_sales)
```

```python
#sales unit Grougth
q2-q1
```

```python
#Percentage Increase or the grouth percentage
(q2-q1)*100/q1
```

```python
#If we need to find the revenue we required price per unit
prices = np.array([
    [10, 12, 11],  # Prices for Product A in different regions
    [8,  9,  10],  # Prices for Product B
    [15, 16, 17]   # Prices for Product C
])

q1_revenue = q1 * prices
q1_revenue
```

```python
#if the product gives discount
# For discount we want to know in each of the region and each of the category what was the discount given in term of dollar($) amount
q1_discount=q1_revenue*0.2
q1_discount
```

```python
#Out of 2000(thousand) 400(Hundred) is the discount
```

```python
# To find Net Revenue then we do revenue-discount=net_revenue
q1_net_revenue=q1_revenue-q1_discount
q1_net_revenue
```

```python
#We want to know the total discount of quarter 1 across the all the products and regions
np.sum(q1_discount)
```

```python
#We want to know the total net_revenue of quarter 1 across the all the products and regions
np.sum(q1_net_revenue)
```

# Dot and Cross Products

## Dot Product
Screenshot 2026-04-17 123820.png

```python
# first column is square footage, second is bedrooms
features = np.array([
    [2000, 3], # house 1
    [1800, 2]  # house 2
])

weights = np.array([150, 50000]) #(2000*150+3*50000=450000,1800*150+2*50000=370000)

price = np.dot(features, weights)
price
```

## Cross Product

```python
c = np.array([1,2,3])
d = np.array([4,5,6])
np.cross(c,d) #|2  3|     |1  3| j + |1  2|
             # |5  6| i - |4  6|     |4  5| k
          #(12-15)i-(6-12)j+(5-8)k
          #-3i-(-6)j-3k
```
